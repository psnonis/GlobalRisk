```python
import numpy                as     np
import pandas               as     pd
import plotly.graph_objects as     go

from   plotly.graph_objects import Bar

risks = pd.read_csv(f'../data/Clean/risks.csv')

x = risks.risk.unique().tolist()
y = risks.groupby('risk').mean().rate.tolist()

bar = Bar(
    x      = y,
    y      = x,
    marker = dict(
        color ='rgba(50, 171, 96, 0.6)',
        line  = dict(
            color = 'rgba(50, 171, 96, 1.0)',
            width = 1
        ),
    ),
    name        = 'Total Risk Profile',
    orientation = 'h',
)

fig = go.Figure(data = bar)

fig.update_layout(
    title = 'Total Risk Profile',
    
    yaxis = dict(
        showgrid       = False,
        showline       = False,
        showticklabels = True,
    ),

    xaxis = dict(
        zeroline       = False,
        showline       = False,
        showticklabels = True,
        showgrid       = True,
    ),

    paper_bgcolor = 'rgb(248, 248, 255)',
    plot_bgcolor  = 'rgb(248, 248, 255)',
)

fig.write_html('2_RiskTotal.html')
```

```python
cries = ['CHN']#,'SAU', 'VEN', 'CHN']
final = risks.date.max()
tempo = risks[risks.code.isin(cries) & risks.date.eq(final)]
#top10 = tempo.groupby('nick')['percent'].nlargest(10)

median = {}

for n, r in enumerate(tempo.nick.unique().tolist(), 1):
    if r not in ['COMPOSITE']:
        m = tempo[tempo.nick.eq(r)].percent.median()
        median[r] = m
    
print(f'Top 10 Median Risks of {", ".join(cries)} for {final.strftime("%Y-%B")}\n')

for n,(r,m) in enumerate(sorted(median.items(), key = lambda i : -i[1])[:10], 1):
    print(f'    {n:>2} {r:<21} {m:6.2f}')
```
