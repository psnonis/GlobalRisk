```python
import altair as ai
import pandas as pd
import numpy  as np

raw   = '../data/Raw'
clean = '../data/Clean'

ai.data_transformers.disable_max_rows()
```

```python
risks_china = pd.read_json(f'{clean}/CountryData_ICRG-Risk_China.json')
risks_clean = pd.read_json(f'{clean}/CountryData_ICRG-Risk.json')
```

```python
risks_clean
```

```python
ai.Chart(risks_china).mark_area().encode(
        ai.X('yearmonth(date):T', axis=ai.Axis(format='%Y', domain=False, tickSize=0)),
        ai.Y('sum(rate):Q', stack='center', axis=None),
        ai.Color('risk:N', scale=ai.Scale(scheme='category20b')
    )
).interactive()
```

```python
from vega_datasets import data

counties = ai.topo_feature(data.us_10m.url, 'counties')
source   = data.unemployment.url

ai.Chart(counties).mark_geoshape().encode(
    color='rate:Q'
).transform_lookup(
    lookup='id',
    from_=ai.LookupData(source, 'id', ['rate'])
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300
)
```

```python
import altair as alt
from vega_datasets import data

# Data generators for the background
sphere = alt.sphere()
graticule = alt.graticule()

# Source of land data
source = alt.topo_feature(data.world_110m.url, 'countries')

# Layering and configuring the components
alt.layer(
    alt.Chart(sphere).mark_geoshape(fill='lightblue'),
    alt.Chart(graticule).mark_geoshape(stroke='white', strokeWidth=0.5),
    alt.Chart(source).mark_geoshape(fill='ForestGreen', stroke='black')
).project(
    'naturalEarth1'
).properties(width=600, height=400).configure_view(stroke=None)
```
