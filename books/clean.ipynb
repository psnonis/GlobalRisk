```python
def add(l1 = '', l2 = '', l3 = '', l4 = '', root = '', nick = '', mult = 1.0, span = 0, node = 0, hiers = {}):
    
    risk   = l4 or l3 or l2 or l1
    level  = 4 if l4 else \
             3 if l3 else \
             2 if l2 else \
             1
    
    path   = ' : '.join(filter(None, [l1, l2, l3, l4]))
    
    type   = 'P' if 'Political' in l2 else \
             'E' if  'Economic' in l2 else \
             'F' if 'Financial' in l2 else \
             ''

    hiers[risk] = dict(risk = risk, nick = nick, root = root, mult = mult, span = span, type = type, level = level, node = node)

    return hiers

hiers = add('Composite Risk Rating'                                                                                   , root = ''                     , nick = 'Total'                , mult = 2.0, span = 200., node =  0, hiers =    {})

hiers = add('Composite Risk Rating', 'Political Risk Rating'                                                          , root = 'Total'                , nick = 'Political'            , mult = 1.0, span = 100., node =  1, hiers = hiers)
hiers = add('Composite Risk Rating',  'Economic Risk Rating'                                                          , root = 'Total'                , nick = 'Economic'             , mult = 1.0, span =  50., node =  2, hiers = hiers)
hiers = add('Composite Risk Rating', 'Financial Risk Rating'                                                          , root = 'Total'                , nick = 'Financial'            , mult = 1.0, span =  50., node =  3, hiers = hiers)

hiers = add('Composite Risk Rating', 'Political Risk Rating',      'Government Stability (A)'                         , root = 'Political'            , nick = 'Government Stability' , mult = 1.0, span =  12., node =  4, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',      'Government Stability (A)', 'Government Unity'     , root = 'Government Stability' , nick = 'Government Unity'     , mult = 1.0, span =   4., node =  5, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',      'Government Stability (A)', 'Legislative Strength' , root = 'Government Stability' , nick = 'Legislative Strength' , mult = 1.0, span =   4., node =  6, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',      'Government Stability (A)', 'Popular Support'      , root = 'Government Stability' , nick = 'Popular Support'      , mult = 1.0, span =   4., node =  7, hiers = hiers)

hiers = add('Composite Risk Rating', 'Political Risk Rating',  'Socioeconomic Conditions (B)'                         , root = 'Political'            , nick = 'Socioeconomic'        , mult = 1.0, span =  12., node =  8, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',  'Socioeconomic Conditions (B)', 'Unemployment'         , root = 'Socioeconomic'        , nick = 'Unemployment'         , mult = 1.0, span =   4., node =  9, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',  'Socioeconomic Conditions (B)', 'Consumer Confidence'  , root = 'Socioeconomic'        , nick = 'Consumer Confidence'  , mult = 1.0, span =   4., node = 10, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',  'Socioeconomic Conditions (B)', 'Poverty'              , root = 'Socioeconomic'        , nick = 'Poverty'              , mult = 1.0, span =   4., node = 11, hiers = hiers)

hiers = add('Composite Risk Rating', 'Political Risk Rating',        'Investment Profile (C)'                         , root = 'Political'            , nick = 'Investment Profile'   , mult = 1.0, span =  12., node = 12, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',        'Investment Profile (C)', 'Contract Viability'   , root = 'Investment Profile'   , nick = 'Contract Viability'   , mult = 1.0, span =   4., node = 13, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',        'Investment Profile (C)', 'Profits Repatriation' , root = 'Investment Profile'   , nick = 'Profits Repatriation' , mult = 1.0, span =   4., node = 14, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',        'Investment Profile (C)', 'Payment Delays'       , root = 'Investment Profile'   , nick = 'Payment Delays'       , mult = 1.0, span =   4., node = 15, hiers = hiers)

hiers = add('Composite Risk Rating', 'Political Risk Rating',         'Internal Conflict (D)'                         , root = 'Political'            , nick = 'Internal Conflict'    , mult = 1.0, span =  12., node = 16, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',         'Internal Conflict (D)', 'Civil War'            , root = 'Internal Conflict'    , nick = 'Civil War'            , mult = 1.0, span =   4., node = 17, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',         'Internal Conflict (D)', 'Terrorism'            , root = 'Internal Conflict'    , nick = 'Terrorism'            , mult = 1.0, span =   4., node = 18, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',         'Internal Conflict (D)', 'Civil Disorder'       , root = 'Internal Conflict'    , nick = 'Civil Disorder'       , mult = 1.0, span =   4., node = 19, hiers = hiers)

hiers = add('Composite Risk Rating', 'Political Risk Rating',         'External Conflict (E)'                         , root = 'Political'            , nick = 'External Conflict'    , mult = 1.0, span =  12., node = 20, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',         'External Conflict (E)', 'War'                  , root = 'External Conflict'    , nick = 'Hot War'              , mult = 1.0, span =   4., node = 21, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',         'External Conflict (E)', 'Cross-border Conflict', root = 'External Conflict'    , nick = 'Cold War'             , mult = 1.0, span =   4., node = 22, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',         'External Conflict (E)', 'Foreign Pressures'    , root = 'External Conflict'    , nick = 'Foreign Pressures'    , mult = 1.0, span =   4., node = 23, hiers = hiers)

hiers = add('Composite Risk Rating', 'Political Risk Rating',                'Corruption (F)'                         , root = 'Political'            , nick = 'Corruption'           , mult = 1.0, span =   6., node = 24, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',      'Military in Politics (G)'                         , root = 'Political'            , nick = 'Military in Politics' , mult = 1.0, span =   6., node = 25, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',        'Religious Tensions (H)'                         , root = 'Political'            , nick = 'Religious Tensions'   , mult = 1.0, span =   6., node = 26, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',               'Law & Order (I)'                         , root = 'Political'            , nick = 'Law & Order'          , mult = 1.0, span =   6., node = 27, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',           'Ethnic Tensions (J)'                         , root = 'Political'            , nick = 'Ethnic Tensions'      , mult = 1.0, span =   6., node = 28, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating', 'Democratic Accountability (K)'                         , root = 'Political'            , nick = 'Democracy'            , mult = 1.0, span =   6., node = 29, hiers = hiers)
hiers = add('Composite Risk Rating', 'Political Risk Rating',       'Bureaucracy Quality (L)'                         , root = 'Political'            , nick = 'Bureaucracy Quality'  , mult = 1.0, span =   4., node = 30, hiers = hiers)

hiers = add('Composite Risk Rating',  'Economic Risk Rating', 'Risk for Per Capita GDP'                               , root =  'Economic'            , nick = 'GDP Per Capita'       , mult = 1.0, span =   5., node = 31, hiers = hiers)
hiers = add('Composite Risk Rating',  'Economic Risk Rating', 'Risk for GDP Growth'                                   , root =  'Economic'            , nick = 'GDP Growth'           , mult = 1.0, span =  10., node = 32, hiers = hiers)
hiers = add('Composite Risk Rating',  'Economic Risk Rating', 'Risk for Inflation'                                    , root =  'Economic'            , nick = 'Inflation'            , mult = 1.0, span =  10., node = 33, hiers = hiers)
hiers = add('Composite Risk Rating',  'Economic Risk Rating', 'Risk for Budget Balance'                               , root =  'Economic'            , nick = 'Budget Balance'       , mult = 1.0, span =  10., node = 34, hiers = hiers)
hiers = add('Composite Risk Rating',  'Economic Risk Rating', 'Risk for Current Account as % of GDP'                  , root =  'Economic'            , nick = 'Current Account % GDP', mult = 1.0, span =  15., node = 35, hiers = hiers)

hiers = add('Composite Risk Rating', 'Financial Risk Rating', 'Risk for Foreign Debt'                                 , root = 'Financial'            , nick = 'Foreign Debt'         , mult = 1.0, span =  10., node = 36, hiers = hiers)
hiers = add('Composite Risk Rating', 'Financial Risk Rating', 'Risk for Debt Service'                                 , root = 'Financial'            , nick = 'Debt Service'         , mult = 1.0, span =  10., node = 37, hiers = hiers)
hiers = add('Composite Risk Rating', 'Financial Risk Rating', 'Risk for Current Account as % of XGS'                  , root = 'Financial'            , nick = 'Current Account % XGS', mult = 1.0, span =  15., node = 38, hiers = hiers)
hiers = add('Composite Risk Rating', 'Financial Risk Rating', 'Risk for International Liquidity'                      , root = 'Financial'            , nick = 'Liquidity'            , mult = 1.0, span =   5., node = 39, hiers = hiers)
hiers = add('Composite Risk Rating', 'Financial Risk Rating', 'Risk for Exchange Rate Stability'                      , root = 'Financial'            , nick = 'Exchange Stability'   , mult = 1.0, span =  10., node = 40, hiers = hiers)

def cleanRisks(countries = []):

    from datetime import datetime

    risks = pd.read_csv(f'{dirty}/CountryData_ICRG-Risk.csv', delimiter = ';')
    flat  = []
    dates = risks.drop(['Country','Variable'], axis = 1).columns.to_list()
    risks = risks.where(pd.notnull(risks), None)
    
    nicks = {}
    
    risks = risks.replace({'Variable' : { 'Repatriation'        : 'Profits Repatriation',
                                          'Popular support'     : 'Popular Support'    ,
                                          'Government Cohesion' : 'Government Unity'   }})
    
    codes = pd.read_csv(f'{dirty}/ISO_CountryCodes.csv')
    codes = dict(zip(codes['common'], codes['alpha-3']))
    
    start = datetime( year = 1980, month = 1, day = 1)

    def getRisk(risk):

        return { 'Repatriation'        : 'Profits Repatriation',
                 'Popular support'     : 'Popular Support'    ,
                 'Government Cohesion' : 'Government Unity'   }.get(risk, risk)

    def getRate(risk, rate):

        return hiers[risk]['mult'] * rate if rate else None

    def getRati(risk, rate):

        return hiers[risk]['span'] - rate if rate else None
    
    def getPerc(risk, rate):
        
        return (hiers[risk]['span'] - rate) / hiers[risk]['span'] * 100 if rate else None

    def getSpan(risk):
        
        return hiers[risk]['span']

    def getType(risk):
        
        return hiers[risk]['type']
    
    def getRoot(risk):
        
        return hiers[risk]['root']
    
    def getNick(risk):
        
        return hiers[risk]['nick']
    
    def getDeep(risk):
        
        return hiers[risk]['level']

    for n, row in risks.iterrows():
        if  not countries or row.Country in countries:
            for i, date in enumerate(dates):
                iso   = codes.get(row.Country, None)
                m_y   = datetime(year  = int(date.split('/')[1]),
                                 month = int(date.split('/')[0]),
                                 day   = 1)
                if  iso and m_y >= start:

                    flat += [{
                             # 'country' : row.Country,
                               'code'    : iso,
                             # 'risk'    : getRisk(row.Variable),
                               'nick'    : getNick(row.Variable),
                               'root'    : getRoot(row.Variable),
                               'rate'    : getRate(row.Variable, row[date]),
                             # 'rati'    : getRati(row.Variable, row[date]),
                               'perc'    : getPerc(row.Variable, row[date]),
                             # 'level'   : getDeep(row.Variable),
                             # 'span'    : getSpan(row.Variable),
                             # 'type'    : getType(row.Variable),
                               'date'    : str(m_y).split(' ')[0]
                            }]

    df = pd.DataFrame(flat)

    df.to_csv( f'{clean}/risks.csv.gz', index  = False, compression = 'gzip')
    
    print('done')    

    return df

risks = cleanRisks()
```

```python
def cleanScope(corporations = ['Apple','Boeing','DuPont','Exxon', 'Walmart']):

    cc = pd.read_csv(f'{dirty}/ISO_CountryCodes.csv')
    df = pd.DataFrame()
    
    for corporation in corporations:
    
        cf = pd.read_csv(f'{dirty}/Orbis_CorporateGroup_{corporation}.csv',
                         na_values = ['n.a.'],
                         thousands = ',',
                         header    = 0,
                         names     = ['id', 'name', 'country', 'city', 'employees', 'assets', 'revenue', 'funds', 'profits'],
                        )

        cf['subsidiaries'] = 1 # for aggregation

        cf[['employees', 'assets', 'revenue', 'funds', 'profits', 'subsidiaries']] = \
        cf[['employees', 'assets', 'revenue', 'funds', 'profits', 'subsidiaries']].fillna(value = 0).astype(np.int32)
        
        cf['code'        ] = cf['country'].replace(cc.set_index('alpha-2')['alpha-3'])
        cf['region'      ] = cf['country'].replace(cc.set_index('alpha-2')['region' ])
        cf['country'     ] = cf['country'].replace(cc.set_index('alpha-2')['common' ])

        cf['corporation' ] = corporation

        df = pd.concat([df,cf])
    
    df.to_csv(f'{clean}/scope.csv', index  = False)
    
    return df

scope = cleanScope()
```

```python
def cleanMandA(corporations = ['Apple','Boeing','DuPont','Exxon', 'Walmart']):
    
    df = pd.DataFrame()
    
    def fix(v, scale = 1000):
        
        v = v.strip().replace(',', '').replace('-', '')
    
        return float(v) * 1000 if v else None
    
    for corporation in corporations:
        
        cf = pd.read_csv(f'{dirty}/CapIQ_MandA_{corporation}.csv',
                         na_values   = ['-'],
                         thousands   = ',',
                         header      = 0,
                         names       = ['transaction', 'announced', 'closed', 'type', 'role', 'target', 'buyer', 'seller', 'rounds', 'round_type', 'valuation_pre', 'valuation_post', 'ammount', 'raised'],
                         parse_dates = ['announced', 'closed'],
                         converters  =
                         {
                             'valuation_pre'  : fix,
                             'valuation_post' : fix,
                             'ammount'        : fix,
                             'raised'         : fix
                         }
                        )

        print(corporation)

        cf[['rounds', 'valuation_pre','valuation_post', 'ammount', 'raised']] = \
        cf[['rounds', 'valuation_pre','valuation_post', 'ammount', 'raised']].fillna(value = 0).astype(np.int32)

        cf['corporation'] = corporation

        df = pd.concat([df,cf])

    df.to_csv(f'{clean}/manda.csv', index  = False)
    
    return df

manda = cleanMandA()
```
